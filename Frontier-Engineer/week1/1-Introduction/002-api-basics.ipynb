# NB-01 — API Basics

**Learning goal:** Make your first call to the Anthropic API. Understand the request structure, how to read the response, how tokens work, and how streaming changes the UX.

**Concepts covered**
- `anthropic.Anthropic()` client initialisation
- `client.messages.create()` — the core API call
- `model`, `max_tokens`, `messages` parameters
- Reading `response.content[0].text`
- Token counts — `response.usage.input_tokens` and `response.usage.output_tokens`
- Streaming with `client.messages.stream()`

## Cell 1 — Setup

```python
from dotenv import load_dotenv
import os
import anthropic

# Load ANTHROPIC_API_KEY from the project-root .env (one folder up from week1/).
load_dotenv(dotenv_path=os.path.join('..', '.env'))

assert os.environ.get('ANTHROPIC_API_KEY'), (
    'ANTHROPIC_API_KEY not found. Add it to the project-root .env file.'
)

# The client reads ANTHROPIC_API_KEY from the environment automatically.
client = anthropic.Anthropic()

MODEL = 'claude-haiku-4-5'   # supports temperature / top_k / top_p (Opus 4.8 / 4.7 do not)
print(f'Client ready. Using model: {MODEL}')
```

## Cell 2 — First API Call

```python
response = client.messages.create(
    model=MODEL,
    max_tokens=256,
    messages=[
        {"role": "user", "content": "What is ICH E2A in pharmaceutical regulation?"}
    ]
)

print(response.content[0].text)
```

## Cell 3 — Inspect Token Usage

```python
print(f"Input tokens  : {response.usage.input_tokens}")
print(f"Output tokens : {response.usage.output_tokens}")
print(f"Total tokens  : {response.usage.input_tokens + response.usage.output_tokens}")
```

## Cell 4 — System Message

```python
response = client.messages.create(
    model=MODEL,
    max_tokens=256,
    system="You are a pharmaceutical regulatory affairs specialist. Answer concisely.",
    messages=[
        {"role": "user", "content": "What does a Type II variation mean under EMA CTR?"}
    ]
)

print(response.content[0].text)
```

## Cell 5 — Streaming

```python
with client.messages.stream(
    model=MODEL,
    max_tokens=256,
    messages=[
        {"role": "user", "content": "Explain what a 483 observation is in three sentences."}
    ]
) as stream:
    for text in stream.text_stream:
        print(text, end="", flush=True)
```

### Key Takeaway

The API call is simple: a `system` message sets the context, a `user` message contains the query, and `max_tokens` controls how much the model can say back. Token counts are your cost meter — log them.
