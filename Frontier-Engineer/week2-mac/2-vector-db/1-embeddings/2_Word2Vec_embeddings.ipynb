This is how Word2Vec works — every word gets converted into a numeric vector like this. We won't dig into whether these specific numbers make sense; the point is just to see the mechanism.

```python

# Word Embeddings
# Parameters:
# - vector_size=10: Dimensionality of the word vectors
# - window=5: Maximum distance between current and predicted word within a sentence
# - min_count=1: Ignores all words with total frequency lower than this
# - workers=4: Number of CPU cores to use for training
def word_embeddings():
    # Word2Vec converts words into numeric vectors based on context
    sentences = [
        ['customer', 'bought', 'smartphone'],
        ['customer', 'bought', 'laptop'],
        ['user', 'ordered', 'tablet'],
    ]
    model = Word2Vec(sentences, vector_size=10, window=5, min_count=1, workers=4)

    print("Vector for 'customer':", model.wv['customer'])
```

```python
print("Word Embeddings:")
word_embeddings()
```
